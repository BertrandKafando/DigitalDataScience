```python
# D. Génération de données et affichage
```

```python
# 1. Utiliser l’aide (help) pour voir comment utiliser la fonction datasets.make_blobs.
help(datasets.make_blobs)
```

```python
# 2. Générez 1000 données de deux variables réparties en 4 groupes
x,y = datasets.make_blobs(n_samples=1000, centers=4, n_features=2)
```

```python
# 3. Utilisez les fonctions figure, scatter, title, xlim, ylim, xlabel, ylabel et show pour afficher les données avec des couleurs correspondant aux classes. Les axes x et y seront dans l’intervalle [-15, 15] et devrons avoir un titre. La figure doit aussi avoir un titre.
```

```python
fig = plt.figure(figsize=(10, 10))
# Définir les limites des axes x et y
plt.xlim(-15, 15)
plt.ylim(-15, 15)

ax = fig.add_subplot()
ax.scatter(x[:, 0], x[:, 1], c=y)

plt.title("Visualisation des données")
plt.xlabel("variable 0")
plt.ylabel("variable 1")

plt.show()
```

```python
# 4. Générez 100 données de deux variables réparties en 2 groupes, puis 500 données de deux variables réparties en 3 groupes. Concaténez (vstack et hstack) les deux jeux de données et les numéros de classe pour fusionner les deux jeux de données. Affichez les trois ensembles avec scatter comme précédemment.
```

```python
#Generation de donner
x_100,y_100 = datasets.make_blobs(n_samples=100, centers=2, n_features=2)
x_500,y_500 = datasets.make_blobs(n_samples=500, centers=3, n_features=2)
```

```python
help(np.vstack)
```

```python
#concatenation
#y_total = 
y_total = np.hstack((y_100+2,y_500))
y_total

#x_total = 
x_total = np.vstack((x_100,x_500))
x_total
 
```

```python
y_500
```

```python
fig = plt.figure()
# Définir les limites des axes x et y


ax = fig.add_subplot()
ax.scatter(x_100[:, 0], x_100[:, 1], c=y_100)
plt.title("Visualisation des _100")
plt.xlabel("variable 0")
plt.ylabel("variable 1")

plt.show()
```

```python
fig = plt.figure()
# Définir les limites des axes x et y


ax = fig.add_subplot()
ax.scatter(x_500[:, 0], x_500[:, 1], c=y_500)
plt.title("Visualisation des _500")
plt.xlabel("variable 0")
plt.ylabel("variable 1")

plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
# Définir les limites des axes x et y
plt.xlim(-15, 15)
plt.ylim(-15, 15)

ax = fig.add_subplot()
ax.scatter(x_total[:, 0], x_total[:, 1], c=y_total)
plt.title("Visualisation des _Total")
plt.xlabel("variable 0")
plt.ylabel("variable 1")

plt.show()
```
